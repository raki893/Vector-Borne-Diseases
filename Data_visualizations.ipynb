## 4.3 Visualizing Combined Symptom Effects

### b. Interaction Effects

```python
import matplotlib.pyplot as plt

symptoms = dataset.drop(columns=['prognosis'])

# Heatmap for interaction effects of two symptoms on a third variable (e.g., severity)
# Assuming 'severity' is a column in the dataset
if 'severity' in dataset.columns:
    interaction_effect = dataset.pivot_table(values='severity', index='sudden_fever', columns='headache')
    plt.figure(figsize=(15, 10))
    sns.heatmap(interaction_effect, cmap='coolwarm')
    plt.title('Interaction Effects of Sudden Fever and Headache on Severity')
    plt.show()

# Contour plot for interaction effects
plt.figure(figsize=(15, 10))
sns.kdeplot(x='sudden_fever', y='headache', hue='prognosis', data=dataset, fill=True)
plt.title('Contour Plot of Interaction Effects')
plt.show()
```
