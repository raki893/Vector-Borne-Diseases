```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.proportion import proportions_ztest
from scipy.stats import ttest_ind

file_path = '/Users/rakibul/Desktop/uni koblenz/winter 202324/Research Lab/code/models/trainn.csv'
dataset = pd.read_csv(file_path)

# Define diseases to compare
disease1 = 'Lyme_disease'
disease2 = 'Zika'

# Define symptom to analyze
symptom = 'sudden_fever'

# Z-Test for Proportions
count1 = dataset[dataset['prognosis'] == disease1][symptom].sum()
count2 = dataset[dataset['prognosis'] == disease2][symptom].sum()
nobs1 = dataset[dataset['prognosis'] == disease1].shape[0]
nobs2 = dataset[dataset['prognosis'] == disease2].shape[0]
z_stat, p_value = proportions_ztest([count1, count2], [nobs1, nobs2])
symptom_columns = [symptom]
mean_severity = dataset.groupby('prognosis')[symptom_columns].mean().reset_index()


# Distribution Plot for Symptom Values
plt.figure(figsize=(14, 10))
sns.kdeplot(data=dataset, x=symptom, hue='prognosis', fill=True, common_norm=False, palette="bright")
#plt.title(f'Distribution of {symptom} Across {disease1} and {disease2}')
plt.xlabel('Symptom Severity')
plt.ylabel('Density')
plt.show()
plt.show()
```
